```python
import yfinance as yf
from pandas_datareader.data import web
import datetime as dt
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as pyplot
%matplotlib inline


```

```python
start_date = dt.datetime(2020,1,1)
end_date = dt.datetime(2023,3,1)

spx = web.DataReader('^SPX', 'stooq', start_date, end_date)
spx.head()

```

```python
spx.info()
```

```python
spx.describe()
```

```python
spx_close = spx['Close'].sort_index(ascending=True)
spx_shortterm_df = pd.DataFrame()
#Calculating 5-8-13 moving day averages for short term
spx_shortterm_df['current_closing'] = spx_close.copy()
spx_shortterm_df['5_rolling_spx'] = spx_close.rolling(window=5).mean()
spx_shortterm_df['8_rolling_spx'] = spx_close.rolling(window=50).mean()
spx_shortterm_df['13_rolling_spx'] = spx_close.rolling(window=100).mean()

spx_shortterm_df .head()



#Plotting 
px.line(spx_close)
fig = px.line(spx_shortterm_df)
fig.show()

```

```python
spx_longterm_df = pd.DataFrame()

spx_longterm_df['current_closing'] = spx_close.copy()
#Calculating 50-100-200 moving day averages for long term
spx_longterm_df['fifty_rolling_spx'] = spx_close.rolling(window=50).mean()
spx_longterm_df['hundred_rolling_spx'] = spx_close.rolling(window=100).mean()
spx_longterm_df['twohundred_rolling_spx'] = spx_close.rolling(window=200).mean()

#Plotting
px.line(spx_close)
fig = px.line(spx_longterm_df)
fig.show()
```
